```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# let us visualize y=x function
x=np.arange(-10,11,0.2)
y=x
print(x)
print(y)
```

```python
plt.plot(y)
```

```python
# let us Visualize constant
x=np.ones(50)*5
y=x
print(x)
print(y)
```

```python
plt.plot(y)
```

```python
# Exponential function
x=np.arange(0,2,0.1)
y=np.exp(x)
print(x)
print(y)
```

```python
# let us visualize exponential function
plt.plot(x,y)
plt.title('Exponential function')
plt.xlabel('x')
plt.ylabel('exp(x)')
plt.show()
```

```python
# y=x**2
x=np.arange(-100,101,0.5)
y=x**2
plt.plot(x,y)
```

```python
x=np.arange(-100,101,0.5)
y=-x**2
plt.plot(x,y)
```
